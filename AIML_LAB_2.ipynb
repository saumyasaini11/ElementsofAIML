```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales1.csv")
```

```python
df=pd.read_csv("supermarket_sales1.csv")
df.info()
#to get all info about data and missing values
```

```python
df.head()
```

```python
df.shape
#to get the shape of the dataframe(rows,columns)
```

```python
df.describe()
#Get the summary statistics(mean,std,min,max,etc.)
```

```python
df.isnull()
#check for missing values
```

```python
df.isnull().sum()
#check for missing values
```

```python
df.dropna(inplace=True)
# to drop/delete rows with missing values (if appropriate)
df.head() # inorder to check the output
```

```python
df.fillna(df['Unit price'].mean(),inplace=True)
#fill missing values with the mean, median, of the column
df.head()
```

```python
#Histograms: Visulalize the distribution of numerical variables
sns.histplot(df['Unit price'])
plt.show()
```

```python
#Box Plots: Identigy outliers and understand the spread of data
sns.boxplot(x='Payment', y='Unit price', data=df)
plt.show()
```

```python
#Scatter Plots: Examins relationships between two numerical variables.
sns.scatterplot(x='cogs',y='Unit price',data=df)
plt.show()
```

```python
#Bar Charts: Visualise the distribution of categorial variables.
sns.countplot(x='Branch',data=df)
plt.show()
```

```python
from inspect import CORO_RUNNING
#Correlation Matrix: Understand the relationships between multiple numerical variables.
af=df[['Unit price','Quantity','Tax 5%', 'cogs','gross income','Rating']]
corr=af.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#Groupby: Analyze data based on different categories.
df.groupby('City')['Quantity'].mean()
```

```python
#Create a pivot table
pd.pivot_table(df,values='Unit price',index='Branch',columns='City')
```
