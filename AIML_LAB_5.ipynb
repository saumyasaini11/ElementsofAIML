```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=(pd.read_csv("supermarket_sales.csv"))
data.info()
```

```python
#1.LINE PLOT()
plt.figure()
plt.plot(data['Unit price'],data['Total'],marker='*')
plt.title('Line Plot')
plt.xlabel('Unit price')
plt.ylabel('Total')
plt.show()
```

```python
# 2. BAR CHART (MATPLOTLIB)
plt.figure()
plt.bar(data['Quantity'],data['cogs'])
plt.title('Bar Chart')
plt.xlabel('Quantity')
plt.ylabel('Total')
plt.show()
```

```python
# 3. Scatterplot
plt.figure()
plt.scatter(data['Quantity'],data['Unit price'])
plt.title('Scatter plot')
plt.xlabel('Quantity')
plt.ylabel('Unit price')
plt.show()
```

```python
# 4. histogran
plt.figure()
plt.hist(data['Quantity'], bins=6)
plt.title('Histogram')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
#5 Box PLot (seaborn)
plt.figure()
sns.boxplot(x=data['Quantity'], y='Tax 5%' , data=data)
plt.title('Box Plot')
plt.xlabel('Quantity')
plt.ylabel('Tax 5%')
plt.show()
```

```python
#6.HEATMAP
af=data[['Unit price', 'Quantity', 'Tax 5%', 'cogs', 'gross income', 'Rating']]
corr=af.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='rocket', annot_kws={"bbox": {"boxstyle": "round,pad=0.3", "fc": "yellow", "ec": "black", "lw": 1}})
plt.show()
```
