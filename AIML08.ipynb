```python
import numpy as np
#numpy is used for numerical computations and here it will handle array crfeation and manipulation
import matplotlib.pyplot as plt
# imports the matplotlib.pyplot module commonly used for creating visualization such as plots and graphs
# it is given the allias plt for brevity
from sklearn.cluster import KMeans
# imports the KMeans class from the sklearn.cluster module, KMeans is used to perform k-means clustering
# an unsupervised machinw learning algorithm for grouping data into clusters
```

```python
data = np.array([
    [1,2], [1.5,1.8], [5,8], [8,8], [1,0.6], [9,11], [8,2],
     [10,2], [9,3], [3,4], [3.5,5], [4.5,5],[3.8,4.8]
])
# a 2d numpy array is created manuallu. each inner list represents a data point with two features
# (coordinates in 2d space). this array will act as the input dataset for clustering
data
```

```python
plt.scatter(data[:,0], data[:,1], s=50)
# a scatter plot is created  using the matplotlib.pyplot module
# data[:,0]  accesses the first column(feature 1) of the array (xcoordinate)
# and data[:,1] accesses the second column (feature2)(y coordinates)
# the parameter s=50 sets the size of the points
plt.title("self generated data")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

```python
Kmeans = KMeans(n_clusters=3, random_state=42)
# create an instance of the KMeans class with the following parameters
# n_clusters = 3: specifies the number of clusters to form (3 clusters)
# random_state=42: sets a fixed seed for reproducibility of results
Kmeans.fit(data)
# fits th eKMeans model to the dataset. it perfomr the clustering process,
# assigning each data point to one of the three clusters
```

```python
centers = Kmeans.cluster_centers_
# retrieves the coordinates of the cluster centers(centroids) as a 2d array. each row corresponds to a centroid
labels = Kmeans.labels_
# retrieves the cl,uster labels for each data point. this is a 1d array where each value represents the cluster to which the corresponding data point belongs
```

```python
plt.scatter(data[:,0], data[:,1], c=labels, cmap='viridis', s=50)
# creates a scatter plot of data points
plt.scatter(centers[:,0], centers[:,1], c='red', s=200, marker='X')
# plots the cluster centers on scatter plot
plt.title("K-Means clustering result (self-generated data)")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```
