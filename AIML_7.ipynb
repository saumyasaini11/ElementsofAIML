```python
#importing necessary libraries
import numpy as np
#Import the numpy library, which is used for numerical operarions in Python
#Here , it is mainly used to generate synthetic data
import pandas as pd
#Import the pandas library
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
np.random.seed(42)
feature_1_class_0 = np.random.normal(2, 1, 100)
feature_2_class_0 = np.random.normal(3, 1, 100) #class0 matlab output 0 aeyga yaha
feature_1_class_1 = np.random.normal(6, 1, 100)
feature_2_class_1 = np.random.normal(5, 1, 100) #1 is the SD, first value is the middle point

feature_class_0 = np.column_stack((feature_1_class_0, feature_2_class_0))
feature_class_1 = np.column_stack((feature_1_class_1, feature_2_class_1))
feature_1_class_1#to print the value
```

```python
labels_class_0 = np.zeros(100)
labels_class_1 = np.ones(100)#replacing 100 values of class 1 target value with 1
features = np.vstack((feature_class_0, feature_class_1))
labels = np.concatenate((labels_class_0, labels_class_1))#concatenate , total 200 values now
data = pd.DataFrame(features, columns=['feature_1', 'feature_2'])
data['target'] = labels
data.head()
data#to print the data
```

```python
data = pd.DataFrame(features, columns=['feature_1', 'feature_2'])
data['target'] = labels
X= data[["feature_1","feature_2"]]
y= data["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(max_depth=3, random_state=42) #nO. of nodes =3 . usse bada nahi hoga (depth)
clf.fit(X_train, y_train) #training
y_pred = clf.predict(X_test)#predicting
accuracy = metrics.accuracy_score(y_test, y_pred) #checking accuracy by comparing the value of y_test and y_pred
print(f"Accuracy:, {accuracy:.2f}") # .2f =. 2 decimal place tak value
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=['feature_1', 'feature_2'], class_names=['class_0', 'class_1'],filled=True )
plt.title("Decision Tree Visualization ")
plt.show()
# gini means how pure or impure at that node
#total no. of sample is 140 jisme class 0 ki value is 69 and class1 ki value is 71
#class=class1 because 71<69
```
